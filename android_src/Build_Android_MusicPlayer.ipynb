# Build Music Player Android APK using Buildozer

This notebook will help you build the APK for the Music Player project without using your local computer's resources.

**Note:** This process takes about 15-20 minutes.

```python
# 1. Install Buildozer and dependencies
!pip install buildozer
!sudo apt update
!sudo apt install -y git zip unzip openjdk-17-jdk python3-pip autoconf libtool pkg-config zlib1g-dev libncurses5-dev libncursesw5-dev libtinfo5 cmake libffi-dev libssl-dev
!pip3 install --user --upgrade Cython==0.29.33 virtualenv
```

## Upload your project files
1.  Click the **Folder icon** on the left sidebar.
2.  Drag and drop your `main.py`, `buildozer.spec`, `requirements.txt`, and any other asset files (like images) into the file area.

```python
# 2. Initialize Buildozer (if you didn't upload buildozer.spec)
# If you uploaded buildozer.spec, SKIP this cell.
# !buildozer init
```

```python
# 3. Build the APK
# This command will ask for confirmation (y/n) sometimes, so we use 'yes' to auto-accept.
!yes | buildozer android debug
```

## Download APK
Once the build is finished, the APK will be in the `bin/` directory.
1.  Refresh the file browser on the left.
2.  Open `bin` folder.
3.  Right-click the `.apk` file and select **Download**.
